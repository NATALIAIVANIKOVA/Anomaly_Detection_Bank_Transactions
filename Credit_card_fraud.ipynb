# Алгоритмы для поиска аномалий, основанные на расстояниях

Сможет ли метод, основанный на расстоянии, найти те самые 1.7% мошеннических транзакций?

Поскольку будут использоваться расстояния между точками, нужно привести все признаки к одному масштабу.

```python
from sklearn.preprocessing import RobustScaler
import numpy as np
```

```python
data_features = data.drop('Class', axis = 1)
```

```python
scaler = RobustScaler()
scaled_data = pd.DataFrame(
    data=scaler.fit_transform(data_features),
    columns=data_features.columns
)
```

```python
data.head(2)
```

```python
from sklearn.base import BaseEstimator
from scipy.spatial.distance import cdist
from sklearn.preprocessing import RobustScaler
```

```python
class DistanceOutliers(BaseEstimator):
    def __init__(self, metric, percentile):
        self.metric = metric
        self.percentile = percentile
    def fit(self, X):
        self.centroid = np.mean(X, axis=0).values.reshape(-1, 1).T
        distances_train = cdist(self.centroid, X, metric=self.metric).reshape(-1)
        self.threshold = np.percentile(distances_train, self.percentile)
    def predict(self, X):
        distances = cdist(self.centroid, X, metric=self.metric).reshape(-1)
        predictions = (distances > self.threshold).astype(int)
        return predictions
```

```python
def anomalies_report(outliers):
    print("Total number of outliers: {}\nPercentage of outliers:   {:.2f}%".format(
            sum(outliers), 100*sum(outliers)/len(outliers)))
```

Сначала попробую использовать Евклидову метрику и найти радиус, внутри которого находятся точки, отображающие благонадежные транзакции. В датасете содержится информация о количестве аномальных (то есть мошеннических транзакций), их 492 или 1.7%. Так гиперпаратметр percentile уже подобран.

```python
euclidian_model = DistanceOutliers(metric='euclidean', percentile=98.3)
euclidian_model.fit(scaled_data)
euclidian_outliers = euclidian_model.predict(scaled_data)
anomalies_report(euclidian_outliers)
```

```python
labeled_data = data.copy()
labeled_data['euclidian_outlier'] = euclidian_outliers
```

То же самое с помощью Манхэттенского расстояния

```python
citiblock_model = DistanceOutliers(metric='cityblock', percentile=98.3)
citiblock_model.fit(scaled_data)
cityblock_outliers = citiblock_model.predict(scaled_data)
anomalies_report(cityblock_outliers)
```

```python
labeled_data = labeled_data.copy()
labeled_data['cityblock_outlier'] = cityblock_outliers
```

Как и следовало ожидать, процент аномальных транзакций вычислен правильно в обоих случаях, но удалось ли алгоритмам правильно разметить аномалии?

```python
labeled_data[(labeled_data['Class'] == 1) & (labeled_data['euclidian_outlier'] == 1)].shape
```

Меньше половины мошеннических транзакций помечены, как аномалии.

```python
labeled_data[(labeled_data['Class'] == 1) & (labeled_data['cityblock_outlier'] == 1)].shape
```

cityblock справился немного лучше, больше правильно распознанных аномалий.

# Поиск аномалий в данных с помощью моделей

```python
from sklearn.ensemble import IsolationForest
```

```python
isolation_forest = IsolationForest(n_estimators=200, contamination = 0.017, max_features=1.0, bootstrap=True)
isolation_forest.fit(scaled_data)
isolation_outliers = isolation_forest.predict(scaled_data)
isolation_outliers = np.array([1 if label == -1 else 0 for label in isolation_outliers])
anomalies_report(isolation_outliers)
```

```python
labeled_data = labeled_data.copy()
labeled_data['isolation_outlier'] = isolation_outliers
```

```python
labeled_data[(labeled_data['Class'] == 1) & (labeled_data['isolation_outlier'] == 1)].shape
```

IsolationForest вроде справился лучше, чем методы, основанные на растоянии - модель нашла гораздо больше реально аномальных транзакций - 325 из 492.

```python
true_labels = labeled_data['Class']
predictions = labeled_data['isolation_outlier']
confusion_matrix = metrics.confusion_matrix(true_labels, predictions)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ["Not fraud", "Fraud"])
cm_display.plot()
plt.show()
```

```python
y_true = labeled_data['Class']
y_pred_io = labeled_data['isolation_outlier']
target_names = ['Not fraud', 'FRAUD']
print(classification_report(y_true, y_pred_io, target_names=target_names))
```

Имеет место более высокие precision и recall, чем у методов, основанных на расстоянии.

```python
matthews_corrcoef(y_true, y_pred_io)
```

```python
from sklearn.svm import OneClassSVM
```

```python
one_class_svm = OneClassSVM(nu=0.017, gamma=0.01)
one_class_svm.fit(scaled_data)
svm_outliers = one_class_svm.predict(scaled_data)
svm_outliers = np.array([1 if label == -1 else 0 for label in svm_outliers])
```

```python
anomalies_report(svm_outliers)
```

```python
labeled_data = labeled_data.copy()
labeled_data['svm_outlier'] = svm_outliers
```

```python
labeled_data[(labeled_data['Class'] == 1) & (labeled_data['svm_outlier'] == 1)].shape
```

```python
true_labels = labeled_data['Class']
predictions = labeled_data['svm_outlier']
confusion_matrix = metrics.confusion_matrix(true_labels, predictions)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ["Not fraud", "Fraud"])
cm_display.plot()
plt.show()
```

```python
y_true = labeled_data['Class']
y_pred_svm = labeled_data['svm_outlier']
target_names = ['Not fraud', 'FRAUD']
print(classification_report(y_true, y_pred_svm, target_names=target_names))
```

```python
matthews_corrcoef(y_true, y_pred_svm)
```

```python
print(f'Matthews correlation coefficient for anomaly detection by Euclidean Distance: {matthews_corrcoef(y_true, y_pred_eu):.4f}')
print(f'Matthews correlation coefficient for anomaly detection by Manhatten Distance: {matthews_corrcoef(y_true, y_pred_co):.4f}')
print(f'Matthews correlation coefficient for anomaly detection by Isolation Forest: {matthews_corrcoef(y_true, y_pred_io):.4f}')
print(f'Matthew correlation coefficient for anomaly detection by SVM: {matthews_corrcoef(y_true, y_pred_svm):.4f}')
```
